```python
import pandas as pd

# Sample EEG data (time series from electrodes)
data = {
    'Participant': ['P1','P1','P2','P2','P3','P3'],
    'Electrode': ['E1','E2','E1','E2','E1','E2'],
    'Brainwave': [12.5, 15.2, 11.8, 14.6, 13.4, 16.1]
}

df = pd.DataFrame(data)

# Display dataset
print("EEG Data")
print(df)

# Mean brainwave value
print("\nMean values:")
print(df.groupby('Participant')['Brainwave'].mean())

# Variance
print("\nVariance:")
print(df.groupby('Participant')['Brainwave'].var())

# Peak value
print("\nPeak value:")
print(df['Brainwave'].max())
```
