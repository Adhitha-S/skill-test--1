```python
import numpy as np

steps = np.array([5000, 7000, 8000, 6000, 9000, 10000, 7500])

print("Total steps:", np.sum(steps))
print("Average steps:", np.mean(steps))
print("Max steps:", np.max(steps))
print("Min steps:", np.min(steps))

threshold = 8000
print("High activity days:", steps[steps > threshold])
```
