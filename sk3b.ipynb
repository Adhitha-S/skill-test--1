```python
import numpy as np

stock1 = np.array([
[100,105,98],
[102,108,99],
[101,107,97],
[103,109,100],
[104,110,101]
])

stock2 = np.array([
[98,104,96],
[100,106,97],
[99,105,95],
[101,107,98],
[102,108,99]
])

sum_matrix = stock1 + stock2
diff_matrix = stock1 - stock2

print("Element-wise Sum:\n", sum_matrix)
print("Element-wise Difference:\n", diff_matrix)
```
