```python
scores = [55, 67, 45, 78, 90, 34, 60, 72, 88, 49,
          53, 61, 70, 82, 39, 47, 66, 74, 58, 91]

avg = sum(scores) / len(scores)
highest = max(scores)
lowest = min(scores)

print("Average:", avg)
print("Highest:", highest)
print("Lowest:", lowest)

# Pass / Fail
for i, s in enumerate(scores):
    if s >= 50:
        print("Student", i+1, "Pass")
    else:
        print("Student", i+1, "Fail")

# Grade Assignment
for s in scores:
    if s >= 85:
        grade = "A"
    elif s >= 70:
        grade = "B"
    elif s >= 60:
        grade = "C"
    elif s >= 50:
        grade = "D"
    else:
        grade = "F"
    print(s, "Grade:", grade)
```
