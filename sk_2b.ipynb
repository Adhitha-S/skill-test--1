```python
import numpy as np

marks = np.array([
[80,75,85],
[60,65,70],
[90,88,92],
[55,58,60],
[70,72,68]
])

print("Average:", np.mean(marks))
print("Highest:", np.max(marks))
print("Lowest:", np.min(marks))

total = np.sum(marks, axis=1)
top_student = np.argmax(total)

print("Top student index:", top_student)
```
